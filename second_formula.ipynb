```python
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk
import pprint
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv("slurs_annotated_data.csv")
```

```python
df_slurs = pd.read_csv("slurs.csv", sep=";")
```

```python
slurs = dict(zip(df_slurs.text, df_slurs.severity_rating))
```

```python
def count_sev_score_lyrics(df_line): 
        
    lst = df_line.split("\n") # split lyrics into lines
    num_lines = len(lst) # number of lines
    token_lines = [word_tokenize(s) for s in lst]
    scores_for_sent = {}
    weighted_sev_score_lyrics = 0  # value for averaged sev_score over the whole lyrics

    for i,line in enumerate(token_lines):
        num_of_words = len(line) # number of words in one line
        sev_score_line = 0  # sev score for one line
        for el in line:
            if el in slurs:
                sev_score_line += float(slurs[el])  
        weighted_sev_score = sev_score_line/num_of_words   # weighted sev score for one line
        scores_for_sent[i] = {"num_of_words": num_of_words, "sev_score_per_line": sev_score_line, "weighted_sev_score": weighted_sev_score}
        weighted_sev_score_lyrics += weighted_sev_score # we sum all the weighted sev scores for each line

    result = weighted_sev_score_lyrics/num_lines # get the intended value (average sev score for lyrics divided by number of lines

    #pprint.pprint(scores_for_sent)
    return result
```

```python
df["formula_2_lyrics"] = df["lyrics"].apply(lambda x: count_sev_score_lyrics(x))
```
